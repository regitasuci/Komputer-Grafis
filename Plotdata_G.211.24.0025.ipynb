```python
"""
DATAPLOT5
"""

import matplotlib.pyplot as plt
import numpy as np

plt.axis([0,140,0,100])
plt.axis('on')
plt.grid(True)

t=[20,40,60,80,100,120] #---time
T=[30,35,43,55,70,85] #---temperature
p=[2,3,4,5.3,7.3,9.6] #--pressure
v=[.6,.58,.54,.46,.35,.2] #---volume

pp=[] #-----list for scaled pressure for plotting
for i in np.arange(0,len(p),1):
      pp.append(p[i]*10) #--scale p by 10

vv=[] #---list for scaled volume for plotting
for i in np.arange(0,len(v),1):
      vv.append(v[i]*100) #--scale volume by 100

plt.plot(t,T,color='r',label='Temperature',marker='o') #----plot temperature
plt.plot(t,pp,color='b',label='Pressure',marker='s') #----plot scaled pressure
plt.plot(t,vv,color='g',label='Volume',marker='d') #----plot scaled volume
plt.legend(loc='upper left')

for y in np.arange(0,100+1,20): #----plot pressure scale values
      a=y/10
      a=str(a) #---convert to string for plotting as text
      plt.text(142,y,a,color='b')

plt.xlabel('Time (hrs)') #label axes
plt.ylabel('Temperature K',color='r')
plt.text(151,65,'Pressure (psi)',rotation=90,color='b')

for y in np.arange(100,-1,-20): #---plot volume scale values
      a=y/100
      a=str(a)
      plt.text(162,y,a,color='g')
      plt.text(159,y+2,'_',color='g')

for y in np.arange(1,99,3):
      plt.text(157,y,'-',color='g')

plt.text(170,65,r'Volume (cm3)',rotation=90,color='g') #----label volume scale

plt.title('Compression Test Results') #-title

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
t = np.array([20, 40, 60, 80, 100, 120])
T = np.array([30, 35, 43.5, 75, 90, 85])        # Temperature (K)
p = np.array([2, 3, 4.5, 5.7, 7.3, 9.6])        # Pressure (Pa)
v = np.array([6., 5.8, 5.4, 4.6, 3.5, 2.])      # Volume (mm)

# Figure and main axis (temperature)
fig, ax1 = plt.subplots()

ax1.set_xlabel('Time (hrs)')
ax1.set_ylabel('Temperature (K)', color='red')
ax1.plot(t, T, 'o-r', label='Temperature')
ax1.tick_params(axis='y', labelcolor='red')
ax1.set_ylim(0, 100)

# Pressure axis
ax2 = ax1.twinx()
ax2.set_ylabel('Pressure (Pa)', color='blue')
ax2.plot(t, p, 's-b', label='Pressure')
ax2.tick_params(axis='y', labelcolor='blue')
ax2.set_ylim(0, 10)

# Volume axis (scaled)
ax3 = ax1.twinx()
ax3.spines.right.set_position(("axes", 1.1))  # Shift third y-axis
ax3.set_ylabel('Volume (mm)', color='green')
ax3.plot(t, v, 'd--g', label='Volume')
ax3.tick_params(axis='y', labelcolor='green')
ax3.set_ylim(0, 10)

# Linear regression for volume vs time
n = len(t)
A = (np.sum(t * v) - np.sum(t) * np.sum(v) / n) / (np.sum(t * t) - (np.sum(t) ** 2) / n)
B = np.mean(v) - A * np.mean(t)
vp = A * t + B

# Plot regression line on ax1 (overlapping with volume)
ax1.plot(t, vp, '-', color='green', linewidth=0.5)

# RMS error
rms = np.sqrt(np.mean((v - vp) ** 2))

# Add regression annotation
ax1.text(60, 28, 'v=At+B', color='green')
ax1.text(65, 18, f'A= {A:.5f}', color='green')
ax1.text(65, 12, f'B= {B:.5f}', color='green')
ax1.text(95, 3, f'RMS error= {rms:.3f}', color='green')

# Add volume value labels
ax3.text(20, v[0], f'{v[0]:.4f}', color='green')
ax3.text(120, v[-1], f'{v[-1]:.4f}', color='green')

# Title and legends
fig.suptitle('Compression Test Results')
fig.legend(loc='upper center', bbox_to_anchor=(0.5, 0.15), ncol=3)

plt.tight_layout()
plt.show()
```
